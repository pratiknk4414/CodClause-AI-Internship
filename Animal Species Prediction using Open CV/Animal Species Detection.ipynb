## Animal Species Prediction using Open CV

### Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from keras.preprocessing import image
from keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions

## ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

##### VGG-16
* It is a convolutional neural network that is 16 layers deep.
* The pretrained network can classify images into 1000 object categories 
such as keyboard, mouse, pencil, and many animals.

### Defining a Function for Prediction

```python
def predict_animal_species(image_path):
    # Load the VGG16 model
    model = VGG16(weights='imagenet')
    
    # Load and preprocess the image
    img = image.load_img(image_path, target_size=(224, 224))
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = preprocess_input(img)
    
    # Make predictions
    predictions = model.predict(img)
    decoded_predictions = decode_predictions(predictions, top=3)[0]
    
    # Return the top predictions
    return decoded_predictions
```

```python
img_path = r"C:\Download\download.jpeg"
```

```python
img = cv2.imread(img_path)
plt.imshow(img) 
```

```python
image_path = img_path
predictions = predict_animal_species(image_path)

print("Top 3 Predictions :")
for _, label, probability in predictions:
    print(f"{label}: {probability * 100:.2f}%")
```
